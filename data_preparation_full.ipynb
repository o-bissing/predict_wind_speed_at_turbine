```python
import pandas as pd
```

## Working on TRAIN dataset

```python
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
# Drop the first two rows using their indices
df_train = df_train.drop(index=[0, 1])

# Verify the result
df_train.head()
```

```python
# Count rows where 'target_feature' is NaN
count_nan_target = df_train['target_feature'].isnull().sum()
print(f"Number of rows with NaN in 'target_feature': {count_nan_target}")
```

```python
# Shape before dropping rows
print(f"DataFrame shape before dropping rows: {df_train.shape}")

# Drop rows where 'target_feature' is NaN
df_train = df_train.dropna(subset=['target_feature'])

# Verify the change
print(f"DataFrame shape after dropping rows: {df_train.shape}")
```

```python
# Check for missing values in each column
missing_values = df_train.isnull().sum()
print(missing_values)
```

```python
# Calculate the percentage of missing values for each column
missing_percentage = (df_train.isnull().sum() / len(df_train)) * 100
print(missing_percentage)
```

Parameter "Hub Height (m)" could be important:
The rotor hub is the component that usually holds the blades and connects them to the main shaft of the wind machine.

```python
# Interpolate NaNs
df_train_cleaned_NaNs = df_train.interpolate(method="pad",limit=None)
```

```python
missing_values = df_train_cleaned_NaNs.isnull().sum()
print(missing_values)
```

```python
# List of columns to drop
columns_to_drop = [
    "Unnamed: 0",
    "Timestamp",
    "training"
]

# Drop the specified columns
df_train_ready = df_train_cleaned_NaNs.drop(columns=columns_to_drop, errors='ignore')

# Display the updated df structure
df_train_ready.head()
```

## Working on TEST dataset

```python
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
# Drop the first two rows using their indices
df_test = df_test.drop(index=[0, 1])

# Verify the result
df_test.head()
```

```python
# Check for missing values in each column
missing_values = df_test.isnull().sum()
print(missing_values)
```

```python
# Calculate the percentage of missing values for each column
missing_percentage = (df_test.isnull().sum() / len(df_test)) * 100
print(missing_percentage)
```

```python
# Convert the column to numeric, with error handling
df_test['Unnamed: 0'] = pd.to_numeric(df_test['Unnamed: 0'], errors='coerce')
```

```python
# Interpolate NaNs
df_test_cleaned_NaNs = df_test.interpolate(method="pad",limit=None)
```

```python
print(df_test_cleaned_NaNs.dtypes)
```

```python
missing_values = df_test_cleaned_NaNs.isnull().sum()
print(missing_values)
```

```python
# List of columns to drop
columns_to_drop = [
    "Timestamp"
]

# Drop the specified columns
df_test_ready = df_test_cleaned_NaNs.drop(columns=columns_to_drop, errors='ignore')

# Display the updated df structure
df_test_ready.head()
```
