```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Working on TRAIN dataset

```python
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
# Drop the first two rows using their indices
df_train = df_train.drop(index=[0, 1])

# Verify the result
df_train.head()
```

```python
# Count rows where 'target_feature' is NaN
count_nan_target = df_train['target_feature'].isnull().sum()
print(f"Number of rows with NaN in 'target_feature': {count_nan_target}")
```

```python
# Shape before dropping rows
print(f"DataFrame shape before dropping rows: {df_train.shape}")

# Drop rows where 'target_feature' is NaN
df_train = df_train.dropna(subset=['target_feature'])

# Verify the change
print(f"DataFrame shape after dropping rows: {df_train.shape}")
```

```python
# Check for missing values in each column
missing_values = df_train.isnull().sum()
print(missing_values)
```

```python
# Calculate the percentage of missing values for each column
missing_percentage = (df_train.isnull().sum() / len(df_train)) * 100
print(missing_percentage)
```

```python
df_metadata = pd.read_csv('metaData.csv')
df_metadata
```

Parameter "Hub Height (m)" could be important:
The rotor hub is the component that usually holds the blades and connects them to the main shaft of the wind machine.

```python
# Kelmasch 3 (.1 data) turbine has around 22% of missing values
from geopy.distance import geodesic

# Extract latitude and longitude for Kelmarsh 1
kelmarsh_1_coords = df_metadata.loc[df_metadata['Title'] == 'Kelmarsh 1', ['Latitude', 'Longitude']].iloc[0].values

# Calculate distances to Kelmarsh 1 for all turbines
df_metadata['Distance to Kelmarsh 1 (km)'] = df_metadata.apply(
    lambda row: geodesic((row['Latitude'], row['Longitude']), kelmarsh_1_coords).kilometers, axis=1
)

# Display the updated dataset with distances
print(df_metadata[['Title', 'Latitude', 'Longitude', 'Distance to Kelmarsh 1 (km)']])
```

```python
# Calculate differences relative to Kelmarsh 1
kelmarsh_1_elevation = df_metadata.loc[df_metadata['Title'] == 'Kelmarsh 1', 'Elevation (m)'].iloc[0]
df_metadata['Elevation Diff. (m)'] = df_metadata['Elevation (m)'] - kelmarsh_1_elevation

# Calculate resulting hub height (Elevation + Hub Height)
df_metadata['Resulting Hub Height (m)'] = df_metadata['Elevation (m)'] + df_metadata['Hub Height (m)']

# Calculate resulting hub height difference relative to Kelmarsh 1
kelmarsh_1_hub_height = df_metadata.loc[df_metadata['Title'] == 'Kelmarsh 1', 'Resulting Hub Height (m)'].iloc[0]
df_metadata['Hub Height Diff. (m)'] = df_metadata['Resulting Hub Height (m)'] - kelmarsh_1_hub_height

# Combine all relevant information into a table
final_table = df_metadata[['Title', 'Distance to Kelmarsh 1 (km)', 'Elevation Diff. (m)',
                    'Hub Height Diff. (m)']]

# Display the table
final_table
```

```python
# List of columns to drop
columns_to_drop = [
    "Wind speed (m/s).1",
    "Wind speed, Standard deviation (m/s).1",
    "Wind direction (°).1",
    "Nacelle position (°).1",
    "Nacelle position, Standard deviation (°).1",
    "Vane position 1+2 (°).1",
    "Power (kW).1",
    "Nacelle ambient temperature (°C).1",
    "Generator RPM (RPM).1",
    "Blade angle (pitch position) A (°).1",
]

# Drop the specified columns
df_train_cleaned = df_train.drop(columns=columns_to_drop, errors='ignore')

# Display the updated df structure
df_train_cleaned.head()
```

```python
df_train_cleaned
```

```python
print(df_train_cleaned.dtypes)
```

```python
# Interpolate NaNs
df_train_cleaned_NaNs = df_train_cleaned.interpolate(method="pad",limit=None)
```

```python
missing_values = df_train_cleaned_NaNs.isnull().sum()
print(missing_values)
```

```python
# List of columns to drop
columns_to_drop = [
    "Unnamed: 0",
    "Timestamp",
    "training"
]

# Drop the specified columns
df_train_ready = df_train_cleaned_NaNs.drop(columns=columns_to_drop, errors='ignore')

# Display the updated df structure
df_train_ready.head()
```
